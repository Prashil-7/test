```python
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],
    [2, 120000, 140000, 160000, 190000],
    [3, 200000, 230000, 260000, 300000],
    [4, 180000, 210000, 240000, 270000],
    [5, 160000, 185000, 205000, 230000],])

print("== zomato sales analysis ==")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(" data shapes", sales_data.shape)
print("samople data for first 3 restaurant \n", sales_data[:3])
```

```python
# total sales per year
print(" totaL sales per year", np.sum(sales_data , axis = 0))
yearly_total = np.sum(sales_data[:,1:], axis= 0)
print("year total ", yearly_total)
print("all total",np.sum([yearly_total]))
```

```python
#min sales per res

min_sales = np.min(sales_data[:, 1:] ,axis=1)
print("min sales",min_sales)

#max slaes per year

max_sales = np.max(sales_data[:, 1:], axis=1)
print("max salsae",max_sales)

#aqvg per res

avg_salss = np.average(sales_data[:, 1:], axis=1)
print("avg", avg_salss)

cummulative_sales = np.cumsum(sales_data[ :, 1:], axis=1)
print("cummulative sales", cummulative_sales) 

plt.Figure(figsize=(10,6))
plt.plot(np.mean(cummulative_sales,axis=0))
plt.title("average cummulative sales acroos all restuarant")
plt.xlabel("years")
plt.ylabel("ssles")
plt.grid(True)
plt.show()
```
